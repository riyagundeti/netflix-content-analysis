```python
# Netflix Content Analysis

import pandas as pd
import matplotlib.pyplot as plt

# Loads the dataset
df = pd.read_csv('netflix_titles.csv')

# Basic exploration
print("Shape:", df.shape)
print("\nContent Types:")
print(df['type'].value_counts())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Analysis 1: Movies vs TV Shows
df['type'].value_counts().plot(kind='bar', color=['blue', 'black'])
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.savefig('analysis1_movies_vs_shows.png')
plt.show()
```

```python
# Analysis 2  Top 10 Countries on Netflix
df['country'].value_counts().head(10).plot(kind='bar', color='red')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('analysis2_top_countries.png')
plt.show()
```

```python
#analysis 3:- titles added over time
df['date_added']= pd.to_datetime(df['date_added'],errors='coerce')
df['year_added'] = df['date_added'].dt.year
print(df['date_added'].dtype)
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
# Count titles per year and plot
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='red')
plt.title('Netflix Titles Added Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.savefig('analysis3_titles_over_time.png')
plt.show()
```

```python
#analysis 4: Top 10 Genres on Netflix
# # Splits each row's genres into a list, then "explode" so each genre gets its own row
genres = df['listed_in'].str.split(', ').explode()
# Now count them properly
top_genres = genres.value_counts().head(10)

top_genres.plot(kind='barh', color='red')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.gca().invert_yaxis()  # so #1 genre shows at the top
plt.tight_layout()
plt.savefig('analysis4_top_genres.png')
plt.show()
```

```python

df = df.dropna(subset=['rating', 'release_year'])

# Map raw ratings into buckets
rating_map = {
    'G': 'Kids', 'TV-Y': 'Kids', 'TV-Y7': 'Kids', 'TV-G': 'Kids',
    'PG': 'Teens', 'TV-PG': 'Teens', 'PG-13': 'Teens', 'TV-14': 'Teens',
    'R': 'Adults', 'TV-MA': 'Adults', 'NC-17': 'Adults'
}
df['maturity'] = df['rating'].map(rating_map)
df = df.dropna(subset=['maturity'])

# Only look at recent-ish years for cleaner chart (older years have very few titles)
recent = df[df['release_year'] >= 2008]

trend = recent.groupby(['release_year', 'maturity']).size().unstack(fill_value=0)

# Convert to percentage per year
trend_pct = trend.div(trend.sum(axis=1), axis=0) * 100

trend_pct.plot(kind='area', stacked=True, figsize=(10,6), colormap='viridis')
plt.title('Content Maturity Mix Over Time (%)')
plt.ylabel('% of titles')
plt.show()
```
