```python
# Netflix Content Analysis

import pandas as pd
import matplotlib.pyplot as plt

# Loads the dataset
df = pd.read_csv('netflix_titles.csv')

# Basic exploration
print("Shape:", df.shape)
print("\nContent Types:")
print(df['type'].value_counts())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Analysis 1: Movies vs TV Shows
df['type'].value_counts().plot(kind='bar', color=['blue', 'black'])
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.savefig('analysis1_movies_vs_shows.png')
plt.show()
```

```python
# Analysis 2  Top 10 Countries on Netflix
df['country'].value_counts().head(10).plot(kind='bar', color='red')
plt.title('Top 10 Countries Producing Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('analysis2_top_countries.png')
plt.show()
```

```python
#analysis 3:- titles added over time
df['date_added']= pd.to_datetime(df['date_added'],errors='coerce')
df['year_added'] = df['date_added'].dt.year
print(df['date_added'].dtype)
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
# Count titles per year and plot
df['year_added'].value_counts().sort_index().plot(kind='line', marker='o', color='red')
plt.title('Netflix Titles Added Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.savefig('analysis3_titles_over_time.png')
plt.show()
```
